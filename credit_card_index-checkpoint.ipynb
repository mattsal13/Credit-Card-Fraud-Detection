```python
## First import the package
from sklearn.model_selection import train_test_split
```

```python
###Separate features
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
## Now split the data
## Have 20% for testing
## Set 40 as the random state
## and stratify the split
X_train, X_test, y_train, y_test=train_test_split(X,y,
                                                  test_size=.2,
                                                  random_state=40,
                                                  shuffle=True,
                                                  stratify=y)
```

```python
#Standardize the X_train
#from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler().fit(X_train)
X_train=scaler.transform(X_train)
```

```python
## import the logistic regression method
from sklearn.linear_model import LogisticRegression
```

```python
## Make a model instance here
Lreg=LogisticRegression()
```

```python
## Fit the model here
Lreg.fit(X_train,y_train)
```

```python
#Standardize the X_test
X_test= scaler.transform(X_test)
```

```python
## get the predictions
y_pred=Lreg.predict(X_test)
y_pred
```

```python
#predict the probabilities
y_pred_prob=Lreg.predict_proba(X_test)
y_pred_prob
```

```python
# now we can get the confusion matrix
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred))
```

```python
# now we can get the accuracy_score
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
# now we can get the classification_report to know the precision, the recall, the f1-score, the support
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Keras. 

from keras import models
from keras import layers
from keras import optimizers
from keras import losses
from keras import metrics

## Define categorical funcion. This prepares entries for tensor analysis. 

def to_categorical(y):
    return tf.keras.utils.to_categorical(y)
```

```python
credit_card_df = pd.read_csv("/Users/Matt/Documents/GitHub/Credit-Card-Fraud-Detection/creditcard.csv")
credit_card_df.head()
```

```python
credit_card_df.value_counts('Class')

```

```python
## Empty model
model = models.Sequential()
```

```python
## Make the layers. ONLY RUN THIS ONCE. Make sure the input_shape is correct matches the number of features.

model.add(layers.Dense(16, activation='relu', input_shape=(30,)))

model.add(layers.Dense(16, activation='relu'))

model.add(layers.Dense(2, activation='softmax'))
```

```python
# Gives a model summary

model.summary()
```

```python
# Here we compile the model. We use binary cross-entropy. 

model.compile(optimizer='rmsprop',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
```

```python
# Features 

X = np.array(credit_card_df.iloc[:,:-1]).reshape(-1, 30)
y = np.array(credit_card_df.iloc[:,-1]).reshape(-1, 1)

## Make sure these are numpy arrays



print(np.shape(X), type(X))
print(np.shape(y), type(y))
```

```python
# Train/test/validation split. 

from sklearn.model_selection import train_test_split

X_train_train,X_val,y_train_train,y_val = train_test_split(X, y,
                                                          test_size=.2,
                                                          shuffle=True,
                                                          stratify=y,
                                                          random_state=440)

print(np.shape(X_train_train), type(X_train_train))
print(np.shape(y_train_train), type(y_train_train))
print(np.shape(X_val), type(X_val))
print(np.shape(y_val), type(y_val))
print('\n\n')
print(np.shape(to_categorical(y_train_train)), type(to_categorical(y_train_train)))
print(np.shape(to_categorical(y_val)), type(to_categorical(y_val)))
```

```python
# Fitting the data. 

history = model.fit(X_train_train,
                        to_categorical(y_train_train),
                        epochs = 100,
                        batch_size = 512,
                        validation_data=(X_val,to_categorical(y_val)))
```
